```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# Perform API Calls
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&appid=" + api_key +"&q="
```

```python
# Lists 
city_list = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

record = 1

print(f"Beginning Data Retrieval")
print(f"----------------------------------------")

# Loop through cities
for city in cities:  
    
    try: 
        response = requests.get(f"{url}{city}").json() 
        city_list.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        city_record = response["name"]
        print(f"Processing Record {record} | {city_record}")
        print(f"{url}&q={city}")
        
        # Record counter
        record = record + 1
        
        # Pause in loop to avoid exceeding rate limit of API
        time.sleep(1)
        
    # If no record found "skip" to next request
    except:
        print("City not found. Skipping...")
    continue
```

```python
# Dictionary
weather_dict = {
    "City": city_list,
    "Cloudiness":cloudiness, 
    "Country":country,
    "Date":date, 
    "Humidity": humidity,
    "Lat":lat, 
    "Lng":lng, 
    "Max Temp": max_temp,
    "Wind Speed":wind_speed
}

# Create a data frame
weather_df = pd.DataFrame(weather_dict)

# Count of weather data values 
weather_df.count()
```

```python
# Display data frame
weather_df.head()
```

```python
# City Latitude vs Max Temperature
plt.scatter(weather_df["Lat"], weather_df["Max Temp"], marker="o", s=12)
plt.title("City Latitude vs. Max Temperature (7/2/2019)")
plt.ylabel("Max. Temperature (F)")
plt.xlabel("Latitude")
plt.grid()

# Save figure
plt.savefig("Max_Temp_vs_Latitude.png")

# Show plot
plt.show()
```

```python
# City Latitude vs Humidity
plt.scatter(weather_df["Lat"], weather_df["Humidity"], marker="o", s=12)
plt.title("City Latitude vs. Humidity (7/2/2019)")
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.grid()

# Save figure
plt.savefig("Latitude_vs_Humidity.png")

# Show plot
plt.show()
```

```python
# City Latitude vs Cloudiness
plt.scatter(weather_df["Lat"], weather_df["Cloudiness"], marker="o", s=12)
plt.title("City Latitude vs. Cloudiness (7/2/2019)")
plt.ylabel("Cloudiness (%)")
plt.xlabel("Latitude")
plt.grid()

# Save figure
plt.savefig("Latitude_vs_Cloudiness.png")

# Show plot
plt.show()
```

```python
# City Latitude vs Wind Speed
plt.scatter(weather_df["Lat"], weather_df["Wind Speed"], marker="o", s=12)
plt.title("City Latitude vs. Wind Speed (7/2/2019)")
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.grid()

# Save figure
plt.savefig("Latitude_vs_Wind.png")

# Show plot
plt.show()
```

# Observations

- As anticipated, temperature increases as we move towards the equator, with temperature peaking around 20 degrees latitude.
- There does not appear to be a correlation between latitude and cloudiness or humidity. 
- Higher speeds of wind appear to be more frequent further away from the equator, however the data does not suggest a strong correlation.
